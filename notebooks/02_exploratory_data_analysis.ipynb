#### grouping for question 1

```python
import pandas as pd
df = pd.read_csv('C:\\Users\\FRANCIS\\Documents\\Operational-Analysis-of-Food-Delivery\\data\\processed\\delivery_master_clean.csv')
grouped_real = df.groupby('Delivery Duration (in minutes)')['Realistic Rating Order Time'].mean().sort_values(ascending=True)
```

#### Plotting for question 1: Find out the exact minute where clients get upset for waiting their food

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(grouped_real.index, grouped_real.values, marker='o')
plt.xlabel('Delivery Duration (in minutes)')
plt.ylabel('Average Rating')
plt.title('Average Rating vs. Delivery Duration')
plt.grid(True)
plt.show()
```

#### Grouping for question 2

```python
grouped_cities = df.groupby('Order_City')['Customer_Rating_Cities'].mean().sort_values(ascending=True)
```

#### Plotting for question 2: Which are the most problematics cities

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.bar(grouped_cities.index, grouped_cities.values, color='tomato')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.title('Average Rating by City')
plt.xticks(rotation=45)
plt.show()
```

#### Grouping for question 4

```python
df['Order_Hour'] = pd.to_datetime(df['Order Date and Time']).dt.hour
grouped_order_time = df.groupby('Order_Hour')['Customer_Rating_Order_Time'].mean().sort_values(ascending=True)
```

#### Plotting for question 4: What time clients are most likely to have less tolerance to higher delivery duration

```python
plt.figure(figsize=(10,6))
plt.plot(grouped_order_time.index, grouped_order_time.values, marker='o', color='royalblue')
plt.xlabel('Order Hour')
plt.ylabel('Average Rating')
plt.title('Average Rating by Order Hour')
plt.grid(True)
plt.xticks(grouped_order_time.index)  
plt.show()
```

#### Plotting for question 5: Find out at if the amount of money spent on the order customers influence in the tolerant of long waits for delivery

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.scatter(df['Total Bill (in Saudi Riyals)'], df['Customer_Rating_Money_Spent'], alpha=0.5, color='teal')
plt.xlabel('Total Bill (in Saudi Riyals)')
plt.ylabel('Customer Rating Money Spent')
plt.title('Customer Rating vs. Total Bill')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df['Bill_Range'] = pd.cut(df['Total Bill (in Saudi Riyals)'], bins=10)
grouped = df.groupby('Bill_Range')['Customer_Rating_Money_Spent'].mean()

plt.figure(figsize=(10,6))
plt.plot(grouped.index.astype(str), grouped.values, marker='o', color='orange')
plt.xlabel('Total Bill Range (in Saudi Riyals)')
plt.ylabel('Average Rating')
plt.title('Average Rating by Total Bill Range')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
