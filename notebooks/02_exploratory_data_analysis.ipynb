#### grouping for question 1

```python
import pandas as pd
df = pd.read_csv('C:\\Users\\FRANCIS\\Documents\\Operational-Analysis-of-Food-Delivery\\data\\processed\\delivery_master_clean.csv')
grouped_real = df.groupby('Delivery Duration (in minutes)')['Realistic Rating Order Time'].mean().sort_values(ascending=True)
```

#### Plotting for question 1: Find out the exact minute where clients get upset for waiting their food

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(grouped_real.index, grouped_real.values, marker='o')
plt.xlabel('Delivery Duration (in minutes)')
plt.ylabel('Average Rating')
plt.title('Average Rating vs. Delivery Duration')
plt.grid(True)
plt.show()
```
